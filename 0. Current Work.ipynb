```python
import pandas as pd
wt = 21.30
wv = 3.52
wl = 25.61
```

```python
clusters = pd.read_csv('./data/stat_cluster.txt',sep='\t')
clusters = clusters.loc[:,["UID","RG","STD_T","STD_V"]]
clusters['RG'] = clusters['RG']/wl
clusters['STD_T'] = clusters['STD_T']/wt
clusters['STD_V'] = clusters['STD_V']/wv
clusters['SUM_ALL'] = clusters['RG']+clusters['STD_V']+clusters['STD_T']
clusters['RC_L'] = clusters['RG']/clusters['SUM_ALL'] 
clusters['RC_T'] = clusters['STD_T']/clusters['SUM_ALL'] 
clusters['RC_V'] = clusters['STD_V']/clusters['SUM_ALL'] 

rcs = clusters.loc[:,['UID','RC_L','RC_T','RC_V']]
rcs['UID'] = rcs['UID'].astype(int)
rcs['TypeL'] = rcs['RC_L'].map(lambda x:'space' if x<0.2 else None)
rcs['TypeT'] = rcs['RC_T'].map(lambda x:'time' if x<0.2 else None)
rcs['TypeV'] = rcs['RC_V'].map(lambda x:'volume' if x<0.2 else None)
# rcs[rcs['TypeV']=='volume']
rcs
```
